```python
from transformers import AutoModelForCausalLM, AutoTokenizer
from recon_trace.tracer import ReconTracer
import torch.nn as nn

model = AutoModelForCausalLM.from_pretrained("gpt2")
tokenizer = AutoTokenizer.from_pretrained("gpt2")

tracer = ReconTracer(model)

# Hook all relevant modules
tracer.add_hooks_all()

inputs = tokenizer("Hello World", return_tensors="pt")["input_ids"]
tracer.run(inputs)
tracer.export_json("activity_cube_all_layers.json")
```
